```python
import numpy as np
import pandas as pd
```

```python
features = pd.read_csv('/Volumes/SanDisk SSD/PhysioNet-Challenge/CODE-15%/exams.csv')
features
```

```python
numerical_features = features[['age', 'nn_predicted_age', 'timey']]
numerical_summary = numerical_features.describe().loc[['mean', 'std']]
numerical_summary.loc['min'] = numerical_features.min()
numerical_summary.loc['max'] = numerical_features.max()

numerical_summary.loc['min', 'timey'] = numerical_features[numerical_features['timey'] > 0]['timey'].min()

numerical_summary = numerical_summary.map(lambda x: round(x, 4))

categorical_features = features[['is_male', '1dAVb', 'RBBB', 'LBBB', 'SB', 'ST', 'AF', 'death', 'normal_ecg']]
categorical_features.loc[:, 'death'] = categorical_features['death'].astype('bool', errors='ignore')
categorical_summary = categorical_features.mean()

death_true_percentage = categorical_features['death'].apply(lambda x: x == True).mean()
death_null_percentage = features['death'].isna().sum() / len(categorical_features)

categorical_summary['death'] = death_true_percentage
categorical_summary['death_null'] = death_null_percentage

categorical_summary = categorical_summary.map(lambda x: round(x, 4))

print("Numerical Features Summary:")
print(numerical_summary)
print("\nCategorical Features Summary (Percentage of True values):")
print(categorical_summary)
```
