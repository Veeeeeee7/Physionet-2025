```python
import numpy as np
import pandas as pd
```

```python
features = pd.read_csv('/Users/victorli/DATA/PTB-XL/physionet.org/files/ptb-xl/1.0.3/ptbxl_database.csv')
features
```

```python
numerical_features = features[['age', 'height', 'weight']]
numerical_features.describe()
```

```python
categorical_features = features[['sex', 'nurse', 'site', 'strat_fold']].astype('category')
categorical_features.describe()

for column in categorical_features.columns:
    print(f"Distribution for column: {column}")
    print(categorical_features[column].value_counts())
    print("\n")
```

```python
categorical_features = features[['device', 'validated_by_human', 'baseline_drift', 'static_noise', 'burst_noise', 'electrodes_problems', 'extra_beats']]
summary = categorical_features.describe(include='all')

combined_summary = pd.DataFrame()

for column in categorical_features.columns:
    print(f"Summary for column: {column}")
    counts = categorical_features[column].apply(lambda x: str(x).split(', ') if pd.notna(x) else []).explode().value_counts(dropna=False)
    percentages = categorical_features[column].apply(lambda x: str(x).split(', ') if pd.notna(x) else []).explode().value_counts(normalize=True, dropna=False)
    summary_df = pd.DataFrame({'Count': counts, 'Percentage': percentages})
    summary_df['Column'] = column
    combined_summary = pd.concat([combined_summary, summary_df])

combined_summary.reset_index(inplace=True)
combined_summary.rename(columns={'index': 'Value'}, inplace=True)
combined_summary
```
